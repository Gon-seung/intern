```python
#영화의 개수, 평점의 개수
import pandas as pd 
import numpy as np 
df1=pd.read_csv('./filter_movie.csv')
df2=pd.read_csv('./filter_rating.csv')
print('영화의 개수 : ', len(df1))
print('평점의 개수 : ', len(df2))
```

```python
#장르의 개수
import pandas as pd 
import numpy as np 
df1=pd.read_csv('./filter_movie.csv')

from ast import literal_eval
genres_count = {}
for i in range(len(df1)):
    genres = df1.loc[i,'genres']
    genres = literal_eval(genres)
    for j in genres:
        if j not in genres_count:
            genres_count.update({j : 1})
        else:
            genres_count[j] += 1
data_gen = pd.DataFrame.from_dict([genres_count])
data_gen = data_gen.rename(index = {0 : 'count'})
data_gen = data_gen.T
data_gen = data_gen.sort_values(by = 'count',ascending= False)
print(data_gen.head(3))
print(data_gen.tail(3))
data_gen.plot.bar()
```

```python
data_gen.plot.pie(y='count', autopct="%.2f",fontsize = 13,figsize=(12, 12))
```

```python
#태그의 개수
import pandas as pd 
import numpy as np 
df1=pd.read_csv('./filter_movie.csv')

from ast import literal_eval
keywords_count = {}
for i in range(len(df1)):
    keywords = df1.loc[i,'keywords']
    keywords = literal_eval(keywords)
    for j in keywords:
        if j not in keywords_count:
            keywords_count.update({j : 1})
        else:
            keywords_count[j] += 1
data_gen1 = pd.DataFrame.from_dict([keywords_count])
data_gen1 = data_gen1.rename(index = {0 : 'count'})
data_gen1 = data_gen1.T
data_gen1 = data_gen1.sort_values(by = 'count',ascending= False)
print(data_gen1.head(3))
print(data_gen1.tail(3))
data_gen1.head(30).plot.bar()
```

```python
#일일 평점 개수
import pandas as pd 
import numpy as np 
df1=pd.read_csv('./filter_rating.csv')
from ast import literal_eval

date_infor = df1.groupby('date').size().reset_index(name='counts')
date_infor = date_infor.sort_values(by = 'counts', ascending= False)
print(date_infor.head(3))
print(date_infor.tail(3))
date_infor.head(30).plot(x = 'date', kind="bar")
```

```python
#각 콜렉션의 개수
import pandas as pd 
import numpy as np 
df1=pd.read_csv('./filter_movie.csv')

collection_infor = df1.groupby('belongs_to_collection').size().reset_index(name='counts')
collection_infor = collection_infor.sort_values(by = 'counts' , ascending= False)
print(collection_infor.head(3))
print(collection_infor.tail(3))
collection_infor.head(30).plot(x = 'belongs_to_collection', kind="bar")
```
